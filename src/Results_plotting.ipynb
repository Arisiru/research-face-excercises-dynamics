```python
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import time
```

```python
labels = ['ResNet-50', 'TPCNN,\nper exercise', 'TPCNN,\nper session', 'MPHNN']
macro_means = [0.06, 0.14, 0.19, 0.97]
weighted_means = [0.05,0.22, 0.30, 0.97]
```

```python
x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, macro_means, width, label='Macro average', color='pink')
rects2 = ax.bar(x + width/2, weighted_means, width, label='Weighted average', color='red')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('F1, 5-fold average')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()


def autolabel(rects):
    """Attach a text label above each bar in *rects*, displaying its height."""
    for rect in rects:
        height = rect.get_height()
        ax.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 3),  # 3 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')


autolabel(rects1)
autolabel(rects2)

fig.tight_layout()

plt.show()
fig.savefig(f'final_results_{int(time.time())}.pdf', format="pdf", dpi=600)    
```
